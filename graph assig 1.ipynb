```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt

x = np.arange(0.1,1*np.pi,0.9)
y = x**2 - np.sin(x)**2 - 4*x + 1 
plt.plot(x,y)
plt.show()
plt.xlabel('x-axis',fontsize=12)
plt.ylabel('y-axis',fontsize=12)
```
